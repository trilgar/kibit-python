Крок 1: Завантажити дані з Excel (pd.read_excel)
Крок 2: Попередня обробка:
    Заповніть пропущені значення:
        У колонці DeliveryCity — заповніть 'Unknown'
        У колонці Quantity — заповніть 1
        У колонці Discount — заповніть 0.0

Крок 3: Додайте обчислювані колонки
    TotalPrice = Quantity * PricePerItem
    DiscountedPrice = TotalPrice * (1 - Discount)
Крок 4: Знайдіть непотрібну колонку та видаліть її

```python
import pandas as pd

orders_df = pd.read_excel('orders.xlsx', index_col=0)
orders_df
```

```python
orders_df =orders_df.fillna({'DeliveryCity': 'Unknown', 'Quantity': 1, 'Discount': 0.0})
orders_df
```

```python
orders_df['TotalPrice'] = orders_df['Quantity'] * orders_df['PricePerItem']
orders_df['DiscountedPrice'] = orders_df['TotalPrice'] * (1 - orders_df['Discount'])
```

```python
del orders_df['CustomersCatName']
```

```python
orders_df.to_excel('modified_orders.xlsx', sheet_name='Customer Orders')
```
