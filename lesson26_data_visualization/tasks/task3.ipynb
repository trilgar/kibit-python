```python
import seaborn as sns, numpy as np
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", palette="Set2")

penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# 0) очистка
penguins = penguins.dropna(subset=["bill_length_mm", "flipper_length_mm", "species", "sex", "body_mass_g"])
penguins.head()
```

```python
ax = sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="flipper_length_mm",
    hue="species",
    style="sex"
)
ax.set(xlabel="Довжина дзьоба, мм", ylabel="Довжина плавця, мм", title="Співвідношення довжини дзьоба і довжини плавця")
plt.show()
```

```python
ax = sns.barplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    hue="sex",
    estimator=np.mean,
    errorbar=None
)
ax.set(xlabel="Вид", ylabel="Середня маса тіла", title="Середня маса тіла пінгвінів")
plt.show()
```
