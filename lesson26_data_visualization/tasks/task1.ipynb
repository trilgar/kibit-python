```python
import numpy as np

rng = np.random.default_rng(42)
# 25 точок: від 0 до 24 год включно
hours = np.linspace(0, 24, 25)

# Модель зовнішньої температури: синусоїда + шум
outdoor = 10 + 8*np.sin(2*np.pi*(hours-6)/24) + rng.normal(0, 0.8, hours.size)
# Модель внутрішньої температури: менша амплітуда, інший зсув фази
indoor  = 21 + 2*np.sin(2*np.pi*(hours-2)/24) + rng.normal(0, 0.4, hours.size)
```

```python
import matplotlib.pyplot as plt
plt.plot(hours, outdoor)
```

```python
plt.plot(hours, outdoor, marker='o', linestyle='-', linewidth=2, markersize = 2)
```

```python
plt.figure()
plt.plot(hours, outdoor, marker='o', linestyle='-', linewidth=2, markersize = 2, label='outdoor')
plt.plot(hours, indoor, marker='o', linestyle='--', linewidth=2, markersize = 2, label='indoor')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(hours, outdoor, marker='o', linestyle='-', linewidth=2, markersize = 2, label='outdoor')
plt.plot(hours, indoor, marker='o', linestyle='--', linewidth=2, markersize = 2, label='indoor')
plt.legend()
plt.title('Temperature')
plt.xlabel('Hour')
plt.ylabel('Temperature, °C')
plt.grid(True)
plt.axis([0, 24, min(outdoor.min(), indoor.min()) , max(outdoor.max(), indoor.max())])
plt.show()
```

```python
plt.figure()
plt.plot(hours, outdoor, marker='o', linestyle='-', linewidth=2, markersize = 2, label='outdoor', alpha=0.1)
plt.plot(hours, indoor, marker='o', linestyle='--', linewidth=2, markersize = 2, label='indoor', color='#1f77b4')
plt.legend()
plt.title('Temperature')
plt.xlabel('Hour')
plt.ylabel('Temperature, °C')
plt.grid(True)
plt.axis([0, 24, min(outdoor.min(), indoor.min()) , max(outdoor.max(), indoor.max())])
plt.savefig('temps.png', dpi=150, bbox_inches='tight')
plt.show()
```
