```python
import pandas as pd

df = pd.DataFrame({
    'year': 2015,
    'month': [1,1,2,2,3,3,4,4],
    'day':   [23,28, 3,21,15,24, 8,24],
    'readings': [12,22,25,20,18,15,17,14]
})
# один стовпець з типом datetime
df['events'] = pd.to_datetime(df[['year','month','day']])
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df['events'], df['readings'])
```

```python
import matplotlib.dates as mdates
fig, ax = plt.subplots()
ax.plot(df['events'], df['readings'], marker='o')

# 1) Де ставити тики
ax.xaxis.set_major_locator(mdates.MonthLocator())   # головні — по місяцях
ax.xaxis.set_minor_locator(mdates.DayLocator())     # другорядні — по днях

# 2) Як їх підписувати
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

# 3) Дрібні покращення
ax.grid(True, which='both', linestyle=':')
fig.autofmt_xdate()  # акуратно повертає підписи
ax.set_ylabel('Readings')
ax.set_title('Readings over time')
```
