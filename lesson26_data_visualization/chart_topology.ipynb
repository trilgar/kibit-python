```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.sin(3*x)/x
plt.plot(x, y)
```

# Histogram

```python
rng = np.random.default_rng(42)
sample = rng.normal(loc=0, scale=1, size=100)
n, edges, patches = plt.hist(sample, bins=10)
print("counts:", n)
print("bin edges:", edges)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1) 
pop = np.random.randint(0, 100, 100)

# 2)  +  
n, bins, patches = plt.hist(pop, bins=20)

# 3) 
plt.title("Histogram of 100 random integers (0100)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# 4) 
plt.show()

# 5)     
print("Counts per bin:", n.astype(int))
print("Bin edges:", bins)
```

# Contour plots

```python
import numpy as np, matplotlib.pyplot as plt

# 1) 
dx = dy = 0.01
x = np.arange(-2.0, 2.0, dx)
y = np.arange(-2.0, 2.0, dy)
X, Y = np.meshgrid(x, y)

# 2) 
def f(x, y):
    return (1 - y**5 + x**5) * np.exp(-x**2 - y**2)

Z = f(X, Y)

# 3)   ()
C = plt.contour(X, Y, Z, 8, colors='black')
plt.clabel(C, inline=1, fontsize=10)   #  
plt.show()
```

```python
levels = np.linspace(-0.8, 0.8, 9)  #    -0.8  0.8
C = plt.contour(X, Y, Z, levels=levels, colors='black', linewidths=1)
plt.clabel(C, fmt="%.2f")
plt.contourf(X, Y, Z, levels=levels)  #    
plt.colorbar()
```

```python
plt.figure()
C = plt.contour(X, Y, Z, 10, colors='black', linewidths=0.8)
plt.clabel(C, inline=True, fontsize=9)
plt.contourf(X, Y, Z, 10, cmap='coolwarm', alpha=0.9)
plt.colorbar()
plt.show()
```

# Polar diagrams

```python
import numpy as np
import matplotlib.pyplot as plt

#  
N = 8

#    (0 .. 2   2/N)
theta = np.arange(0.0, 2*np.pi, 2*np.pi / N)

#   ()
radii = np.array([4, 7, 5, 3, 1, 5, 6, 7])

#   (hex)
colors = np.array([
    '#4bb2c5', '#c5b47f', '#EAA228', '#579575',
    '#839557', '#958c12', '#953579', '#4b5de4'
])

#   (   )
plt.axes([0.025, 0.025, 0.95, 0.95], polar=True)

#  ""   
bars = plt.bar(theta, radii, width=(2*np.pi / N), bottom=0.0, color=colors)

plt.show()
```

# 3d charts

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

X = np.arange(-2, 2, 0.1)
Y = np.arange(-2, 2, 0.1)
X, Y = np.meshgrid(X, Y)

def f(x, y):
    return (1 - y**5 + x**5) * np.exp(-x**2 - y**2)

ax.plot_surface(X, Y, f(X, Y), rstride=1, cstride=1)
ax.view_init(elev=50,azim=250)
plt.show()
```
