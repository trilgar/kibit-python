```python
import pandas as pd

data = {
    "type":   ['A','B','C','D','E'],
    "series1":[1,3,4,3,5],
    "series2":[2,4,5,2,4],
    "series3":[3,2,3,1,3],
}
df = pd.DataFrame(data)
df
```

```python
df_unpivot = pd.melt(
    df,
    id_vars="type",
    value_vars=["series1", "series2", "series3"],
    var_name="variable",
    value_name="value",
)
df_unpivot
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

ax = sns.barplot(
    data=df_unpivot, x="type", y="value", hue="variable"
)
ax.set(xlabel="type", ylabel="value", title="Multiseries barplot (Seaborn)")
```
