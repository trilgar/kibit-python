```python
import numpy as np

t  = np.arange(0, 2.5, 0.1)
y1 = np.sin(np.pi * t)                 # фаза 0
y2 = np.sin(np.pi * t + np.pi/2)       # +π/2
y3 = np.sin(np.pi * t - np.pi/2)       # -π/2

plt.plot(t, y1, 'b*',  t, y2, 'g^',  t, y3, 'ys')  # маркери: *, ^, s
plt.title('Синусоїди з фазовими зсувами (маркери)')
plt.grid(True)
plt.show()
```

```python
t  = np.arange(0, 2.5, 0.01)
y1 = np.sin(np.pi * t)
y2 = np.sin(np.pi * t + np.pi/2)
y3 = np.sin(np.pi * t - np.pi/2)

plt.plot(t, y1, 'b--', label='фаза 0')
plt.plot(t, y2, 'g-',  label='+π/2')
plt.plot(t, y3, 'r-.', label='-π/2')
plt.legend()
plt.grid(True)
plt.title('Синусоїди з різними лінійними стилями')
plt.xlabel('t')
plt.ylabel('sin(πt + φ)')
plt.show()
```

```python
x = np.arange(1, 10)
y = np.array([1, 2, 4, 2, 1, 0, 1, 2, 4])

plt.plot(x, y, marker='o', linestyle='-', linewidth=2.0,
         markersize=6, alpha=0.9, label='приклад kwargs')
plt.legend()
plt.title('Приклад використання kwargs')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t  = np.arange(0, 5, 0.01)
y1 = np.sin(2 * np.pi * t)
y2 = np.cos(2 * np.pi * t)

plt.subplot(2, 1, 1)
plt.plot(t, y1, 'b-.')
plt.title('Підграфік 1: sin')

plt.subplot(2, 1, 2)
plt.plot(t, y2, 'r--')
plt.title('Підграфік 2: cos')
plt.tight_layout()
plt.show()
```

```python
t  = np.arange(0, 1, 0.005)
y1 = np.sin(2 * np.pi * t)
y2 = np.cos(2 * np.pi * t)

plt.subplot(1, 2, 1)
plt.plot(t, y1, 'b-.')
plt.title('sin')

plt.subplot(1, 2, 2)
plt.plot(t, y2, 'r--')
plt.title('cos')

plt.tight_layout()
plt.show()
```
