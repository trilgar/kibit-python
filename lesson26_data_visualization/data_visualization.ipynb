```python
import numpy as np

t  = np.arange(0, 2.5, 0.1)
y1 = np.sin(np.pi * t)                 # фаза 0
y2 = np.sin(np.pi * t + np.pi/2)       # +π/2
y3 = np.sin(np.pi * t - np.pi/2)       # -π/2

plt.plot(t, y1, 'b*',  t, y2, 'g^',  t, y3, 'ys')  # маркери: *, ^, s
plt.title('Синусоїди з фазовими зсувами (маркери)')
plt.grid(True)
plt.show()
```

```python
t  = np.arange(0, 2.5, 0.01)
y1 = np.sin(np.pi * t)
y2 = np.sin(np.pi * t + np.pi/2)
y3 = np.sin(np.pi * t - np.pi/2)

plt.plot(t, y1, 'b--', label='фаза 0')
plt.plot(t, y2, 'g-',  label='+π/2')
plt.plot(t, y3, 'r-.', label='-π/2')
plt.legend()
plt.grid(True)
plt.title('Синусоїди з різними лінійними стилями')
plt.xlabel('t')
plt.ylabel('sin(πt + φ)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t  = np.arange(0, 5, 0.01)
y1 = np.sin(2 * np.pi * t)
y2 = np.cos(2 * np.pi * t)

plt.subplot(2, 1, 1)
plt.plot(t, y1, 'b-.')
plt.title('Підграфік 1: sin')

plt.subplot(2, 1, 2)
plt.plot(t, y2, 'r--')
plt.title('Підграфік 2: cos')
plt.tight_layout()
plt.show()
```

```python
t  = np.arange(0, 1, 0.005)
y1 = np.sin(2 * np.pi * t)
y2 = np.cos(2 * np.pi * t)

plt.subplot(1, 2, 1)
plt.plot(t, y1, 'b-.')
plt.title('sin')

plt.subplot(1, 2, 2)
plt.plot(t, y2, 'r--')
plt.title('cos')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4])
y = np.array([1,4,9,16])

plt.plot(x, y, 'ro')
plt.axis([0, 5, 0, 20])
plt.title('My first plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.axis([0, 5, 0, 20])
plt.title("My first plot")
plt.xlabel("Counting")
plt.ylabel("Square values")
plt.plot(x, y, "ro")
plt.show()
```

```python
plt.axis([0, 5, 0, 20])
plt.title("My first plot", fontsize=20, fontname="Times New Roman")
plt.xlabel("Counting", color="gray")
plt.ylabel("Square values", color="gray")
plt.plot([1,2,3,4], [1,4,9,16], "ro")
plt.show()
```

```python
plt.axis([0, 5, 0, 20])
plt.title("My first plot", fontsize=20, fontname="Times New Roman")
plt.xlabel("Counting", color="gray")
plt.ylabel("Square values", color="gray")

x = [1, 2, 3, 4]; y = [1, 4, 9, 16]
labels = ["First", "Second", "Third", "Fourth"]

plt.plot(x, y, "ro")
for xi, yi, lab in zip(x, y, labels):
    plt.text(xi, yi, lab)        # координати ті самі, що й у даних

plt.show()
```

```python
plt.axis([0, 5, 0, 20])
plt.plot([1,2,3,4], [1,4,9,16], "ro")

plt.text(
    1.1, 12, r"$y = x^2$", fontsize=20,
    bbox={"facecolor": "yellow", "alpha": 0.2}
)

plt.show()
```

```python
plt.axis([0, 5, 0, 20])
plt.plot([1,2,3,4], [1,4,9,16], "ro")
plt.grid(True)  # або plt.grid(True, which="both", alpha=0.3)
plt.show()
```

```python
plt.axis([0, 5, 0, 20])
plt.plot([1,2,3,4], [1,4,9,16], "ro")
plt.minorticks_on()
plt.grid(True, which="both", alpha=0.3, linestyle="-.", linewidth=1)  # або plt.grid(True, which="both", alpha=0.3)
plt.show()
```

```python
plt.axis([0, 5, 0, 20])
plt.plot([1,2,3,4], [1,4,9,16], "ro")
plt.legend(["First series"], loc="upper left")
plt.grid(True)
plt.show()
```

```python
# спосіб 1: список підписів
t = [1,2,3,4]
y1 = [1,4,9,16]
y2 = [0.8, 3.5, 8, 15]
y3 = [0.5, 2.5, 4, 12]

plt.plot(t, y1, "ro")
plt.plot(t, y2, "g^")
plt.plot(t, y3, "b*")
plt.legend(["First series", "Second series", "Third series"], loc="upper left")
plt.show()

# спосіб 2: через label=
plt.plot(t, y1, "ro", label="First series")
plt.plot(t, y2, "g^", label="Second series")
plt.plot(t, y3, "b*", label="Third series")
plt.legend(loc="upper left")
plt.show()
```
