```python
import matplotlib.pyplot as plt

plt.figure()
plt.axis([0, 5, 0, 20])
plt.title('My first plot', fontsize=14)
plt.xlabel('Counting', color='gray')
plt.ylabel('Square values', color='gray')
plt.grid(True)

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro', label='First series')
plt.legend(loc=2)

# ЗБЕРЕЖЕННЯ — має бути НА КІНЦІ!
plt.savefig('my_chart.png', dpi=150, bbox_inches='tight', facecolor='white')
```
