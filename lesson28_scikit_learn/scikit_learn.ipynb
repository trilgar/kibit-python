```python

import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X, y = iris.data, iris.target
names = iris.target_names

print(X.shape)
print(names)
print(X[:3])
print(y[:10])
```

```python
x, y_axis = X[:, 0], X[:, 1]  # sepal length, sepal width
plt.figure()
plt.title('Iris: sepal length vs width')
scatter = plt.scatter(x, y_axis, c=iris.target, cmap='viridis')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(x.min() - .5, x.max() + .5)
plt.ylim(y_axis.min() - .5, y_axis.max() + .5)
plt.legend(handles=scatter.legend_elements()[0], labels=names, loc='best')
plt.show()
```

```python
x = iris.data[:, 2]  # petal length
y = iris.data[:, 3]  # petal width
species = iris.target

#    
x_min, x_max = x.min() - .5, x.max() + .5
y_min, y_max = y.min() - .5, y.max() + .5

plt.figure()
plt.title('Iris Dataset  Classification by Petal Sizes', fontsize=14)
plt.scatter(x, y, c=species)
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np

covar_matrix = PCA(n_components=4)
covar_matrix.fit(iris.data)
variance = covar_matrix.explained_variance_ratio_
variance
```

```python
cum_var = np.cumsum(variance) * 100
```

```python
plt.figure()
plt.plot([1,2,3,4], cum_var, marker='o')
plt.axvline(3, ls='--', color='#bbbbbb')
plt.ylim(90, 100.5)
plt.xticks([1,2,3,4])
plt.xlabel('# of features (PC)')
plt.ylabel('% variance explained')
plt.title('PCA Analysis (Iris)')
plt.show()
```

```python
X = iris.data
y = iris.target
X_std = StandardScaler().fit_transform(X)
X_red = PCA(n_components=3).fit_transform(X_std)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_red[:,0], X_red[:,1], X_red[:,2], c=y, cmap='viridis')
ax.set_title('Iris by PCA (3D)')
ax.set_xlabel('PC1'); ax.set_ylabel('PC2'); ax.set_zlabel('PC3')
plt.show()
```

```python
X = iris.data
y = iris.target
X_std = StandardScaler().fit_transform(X)
X_red_2d = PCA(n_components=2).fit_transform(X_std)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X_red_2d[:,0], X_red_2d[:,1], c=y, cmap='viridis')
ax.set_title('Iris by PCA (2D)')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
plt.show()
```

# K nearest neighbors

```python
np.random.seed(0)
iris = datasets.load_iris()
X = iris.data
y = iris.target

indices = np.random.permutation(len(iris.data))
X_train = X[indices[:-10]]
y_train = y[indices[:-10]]
X_test = X[indices[-10:]]
y_test = y[indices[-10:]]
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
print(y_pred)
print(y_test)
```

```python
from matplotlib.colors import ListedColormap
iris = datasets.load_iris()
X = iris.data[:, :2]  #     
y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = .02

xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

cmap_light = ListedColormap(['#AAAAFF', '#AAFFAA', '#FFAAAA'])
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)

Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light, shading='auto')
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('KNN decision boundaries')
plt.show()
```
