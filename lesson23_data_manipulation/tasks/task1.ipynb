```python
import numpy as np
import pandas as pd

# 1. вихідні дані
items = pd.DataFrame({
    "id":    ["ball", "pen", "mug"],
    "color": ["white", "black", "red"],
    "price": [12.00,  2.50,   7.80]
})

sales_q1 = pd.DataFrame({
    "id":    ["ball", "pen", "pen", "mug", "ball"],
    "month": ["Jan",  "Jan", "Feb",  "Feb", "Mar"],
    "qty":   [10,      35,    25,     12,    18]
})
sales_q2 = pd.DataFrame({
    "id":    ["ball", "pen", "mug", "mug", "pen"],
    "month": ["Apr",  "Apr", "May", "May", "Jun"],
    "qty":   [20,      22,    15,    19,    14]
})
```

```python
sales_q1
```

```python
sales_q2
```

```python
# 2. обʼєднання кварталів
sales = pd.concat([sales_q1, sales_q2], ignore_index=True)
sales
```

```python
# 3. приєднання прайс-листа і виручка
merged = pd.merge(sales, items, on="id", how="left")
merged
```

```python
merged["revenue"] = merged["qty"] * merged["price"]
merged
```

```python
rev_pivot = merged.pivot_table(
    index="month", columns="id", values="revenue", aggfunc="sum"
)
rev_pivot
```

```python
long  = rev_pivot.stack()
long
```

```python
wide  = long.unstack()
wide
```
